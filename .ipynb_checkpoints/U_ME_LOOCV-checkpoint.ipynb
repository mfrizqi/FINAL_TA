```python
import pandas as pd
import numpy as np 
import joblib
import matplotlib.pyplot as plt

from sklearn import model_selection
from sklearn import metrics
from sklearn.svm import SVR
from sklearn.metrics import r2_score
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
param_grid={
            'C': 10,
            'gamma': 'scale',
            'epsilon':0.1
}

c = param_grid['C']
gam = param_grid['gamma']
eps = param_grid['epsilon']
```

```python
data = pd.read_csv("data/all_data.csv")
desc_25 = joblib.load('./pickle/trial_10/bestDesc_25_2050.pkl')
```

```python
from sklearn.model_selection import LeaveOneOut 
X = data.loc[:,desc_25]
y = data.iloc[:, [-1]]
loo = LeaveOneOut()
loo.get_n_splits(X)
```

```python
r2_scores = pd.DataFrame(columns=['r2_train','r2_test','train_pred','test_pred'])

for train_index, test_index in loo.split(X):
    x_train, x_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Feature scaler Using MinMaxScaler()
    scaler = MinMaxScaler()
    scaler.fit(x_train)
    scale_x_train = scaler.transform(x_train)
    scale_x_test = scaler.transform(x_test)
    
    model = SVR(kernel="rbf", C= c, gamma= gam, epsilon=eps)
    model.fit(scale_x_train, y_train)
    
    #Calculate prediction
    y_train_pred = model.predict(scale_x_train)
    y_test_pred = model.predict(scale_x_test)
    
    #Calculate r2 score
    r2_train = r2_score(y_train,y_train_pred)
    r2_test = r2_score(y_test,y_test_pred)
    
    r2_scores = r2_scores.append(pd.Series([r2_train, r2_test, y_train_pred, y_test_pred], index=r2_scores.columns ), ignore_index=True)
```

```python
model = SVR(kernel="rbf", C= c, gamma= gam, epsilon=eps)
scores = cross_val_score(model , X = X , y = y , cv = X.shape[0]) 
scores
```
