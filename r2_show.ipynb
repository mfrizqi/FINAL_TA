```python
import joblib
import matplotlib.pyplot as plt
import numpy as np
```

# SVR

```python
data_rbf = joblib.load('./pickle/with_temp/r2_rbf_neue.pkl')
data_rbf
```

```python
data_linear = joblib.load('./pickle/with_temp/r2_linear_neue.pkl')
data_linear
```

```python
data_poly = joblib.load('./pickle/with_temp/r2_poly_neue.pkl')
data_poly
```

```python
### plot
# x = np.arange(5,26,5)
x = [5,10,15,20,25]

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots(figsize=(8, 6), dpi=80)  # Create a figure and an axes.

ax.plot(x, data_rbf.loc[:,'r2_train'], label='RBF Train',marker='.',markersize=10, lw=2, color='b')
ax.plot(x, data_rbf.loc[:,'r2_test'], label='RBF Val', marker='.', markersize=10, lw=2, ls='dotted', color='b') 
ax.plot(x, data_linear.loc[:,'r2_train'], label='Linear Train', marker='.',markersize=10, lw=2,color='r')
ax.plot(x, data_linear.loc[:,'r2_test'], label='Linear Val', marker='.', markersize=10, lw=2, ls='dotted',color='r')
ax.plot(x, data_poly.loc[:,'r2_train'], label='Poly Train', marker='.', markersize=10, lw=2, color='g')
ax.plot(x, data_poly.loc[:,'r2_test'], label='Poly Val', marker='.', markersize=10, lw=2, ls='dotted', color='g')

ax.set_xlabel('Number of Descriptors')
ax.set_ylabel('R2 Score')
# ax.set_title("SVR r2 score")  
ax.set_xticks(x)
plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.savefig('./r2_models_numdesc_neue.png', format='png', dpi=1000, bbox_inches="tight")    
plt.show()
```

```python
### plot
# x = np.arange(5,26,5)
x = [5,10,15,20,25]

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots( figsize=(8, 6), dpi=80)  # Create a figure and an axes.

ax.plot(x, NuRBF.loc[:,'r2_train'], label='RBF Train', marker='.',markersize=10, lw=2, color='b')
ax.plot(x, NuRBF.loc[:,'r2_test'], label='RBF Val', marker='.', markersize=10, lw=2, ls='--', color='b') 
ax.plot(x, NuLinear.loc[:,'r2_train'], label='Linear Train', marker='.',markersize=10, lw=2,color='r')
ax.plot(x, NuLinear.loc[:,'r2_test'], label='Linear Val', marker='.', markersize=10, lw=2, ls='--',color='r')
ax.plot(x, NuPoly.loc[:,'r2_train'], label='Poly Train', marker='.', markersize=10, lw=2, color='g')
ax.plot(x, NuPoly.loc[:,'r2_test'], label='Poly Val', marker='.', markersize=10, lw=2, ls='--', color='g')

ax.set_xlabel('Descriptors')
ax.set_ylabel('R2 Score')
ax.set_title("NuSVR")  
ax.set_xticks(x)
plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```
