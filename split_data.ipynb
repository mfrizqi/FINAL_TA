```python
import csv 
import pandas as pd
import numpy as np
```

```python
# with pandas
dfPadel = pd.read_csv("data/padel_desc.csv")
dtRawIC50 = pd.read_csv("data/pIC50_target.csv")
dtIC50 = dtRawIC50['pIC50']
```

```python
# Concate all dataframe (Mordred,Padel,Cdk,Target)
mixData = pd.concat([dfPadel,dtIC50], axis=1)
mixData = mixData.drop(columns = ['Name'])
mixData
```

```python
# Export concatenated data
mixData.to_csv (r'data\all_data.csv',index=False, header=True,sep=',')
```

```python
### Run This if you done FS with all_data.csv
from sklearn.model_selection import train_test_split
df = pd.read_csv("data/all_data.csv", delimiter=',')

train,test = train_test_split(df,test_size=0.20, random_state=45)
train.shape, test.shape
```
