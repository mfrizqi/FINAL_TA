```python
import csv 
import pandas as pd
import numpy as np
```

```python
# with pandas
dfPadel = pd.read_csv("data/padel_desc.csv")
dtRawIC50 = pd.read_csv("data/pIC50_target.csv")
dtIC50 = dtRawIC50['pIC50']
```

```python
# Concate all dataframe (Mordred,Padel,Cdk,Target)
mixData = pd.concat([dfPadel,dtIC50], axis=1)
mixData = mixData.drop(columns = ['Name'])

# Export concatenated data
mixData.to_csv (r'data\all_data.csv',index=False, header=True,sep=',')
mixData.head()
```

```python
df = mixData
train=df.sample(frac=0.8,random_state=45) #random state is a seed value
test=df.drop(train.index)
train.index, test.index
```

```python
train.to_pickle("./pickle/train_10.pkl")
test.to_pickle("./pickle/test_10.pkl")
```
